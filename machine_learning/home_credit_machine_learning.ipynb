```python
# import libreries
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import psycopg2 as pg
```

```python
# extract data application_test
application_test = pd.read_sql_query('SELECT * FROM application_test', con=engine)
application_test.head()
```

```python
# extract data application_train
application_train = pd.read_sql_query('SELECT * FROM application_train', con=engine)
application_train.head()
```

DATA CLEANSING

```python
# drop columns if null value is more than 60 from application_train

drop_columns = []
for col in application_train.columns:
    null_pct = application_train[col].isna().sum() / len(application_train)
    if null_pct > 0.06:
        drop_columns.append(col)

application_train_clean = application_train.drop(drop_columns, axis=1)
application_train_clean.head()
```

```python
# drop columns if null value is more than 60 from application_test

drop_columns = []
for col in application_test.columns:
    null_pct = application_test[col].isna().sum() / len(application_test)
    if null_pct > 0.06:
        drop_columns.append(col)

application_test_clean = application_test.drop(drop_columns, axis=1)
application_test_clean.head()
```

```python
# Export data to csv
application_train_clean.to_csv('/home/fajarsetia/Documents/de9-final-project-main/machine_learning/data-csv/application_train_clean.csv', index=False)
application_test_clean.to_csv('/home/fajarsetia/Documents/de9-final-project-main/machine_learning/data-csv/application_test_clean.csv', index=False)
```

```python
# Create  table in postgres
application_test_clean.to_sql('application_test_clean',con=engine, if_exists='replace')
```

```python
# Create  table in postgres
application_train_clean.to_sql('application_train_clean',con=engine, if_exists='replace')
```

Preprocessing

```python
# Splitting data
X_train = application_train_clean.drop(columns=['TARGET'],axis=1)
y_train = application_train_clean['TARGET']
X_test = application_test_clean
```

```python
# Numerical Columns
num=X_train.select_dtypes(exclude='object').columns

# Categorical Columns
cat=X_train.select_dtypes(include='object').columns
```

Logistic Regression Modelling

```python
# import libraries
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LogisticRegression
from imblearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

from sklearn.metrics import precision_recall_fscore_support
```

```python
# Transformer and Modelling Steps
cat_transformer = Pipeline([
    ("c_i", SimpleImputer(strategy="most_frequent")),
    ("e", OneHotEncoder(handle_unknown = "ignore"))
])

num_transformer = Pipeline([
    ("n_i", SimpleImputer(strategy="median"))
])

transformers = [
    ("n_t", num_transformer, num),
    ("c_t", cat_transformer, cat)
]

model = Pipeline([
    ("pre", ColumnTransformer(transformers=transformers)),
    ("model", LogisticRegression(max_iter=1000))
])
```

```python
# Fitting model
model.fit(X_train, y_train)
```

```python
# Predict y_test

y_test = model.predict(X_test)
```

```python
# Prediction Probability
y_prob = model.predict_proba(X_test)
```

Upload ML Result to Postgres

```python
# combine the result of machine learning

# get id
result = X_test[['SK_ID_CURR']]

# add ml result
result['prediction_target'] = y_test.tolist()
result['probability'] = y_prob.tolist()

result
```

```python
# upload ml result to postgres
result.to_sql('home_credit_result_ML', index=False, con=engine, if_exists='replace')
```
